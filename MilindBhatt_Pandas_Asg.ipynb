#**30. Perform the following operations on the dataframes given below.**
A = pd.DataFrame([['Carl', 22],['Martha', 25],['Calvin', 12],['Stuart', 15]], columns=[â€˜Nameâ€™, â€˜Ageâ€™])    
B = pd.DataFrame([['Melvin', 25],['Martha', 34],['Lewis', 32],['Leo', 25]], columns=[â€˜Nameâ€™, â€˜Ageâ€™])
1. Left Outer Join
2. Outer Join
3. Inner Join
4. Right Outer Join

```python
A = pd.DataFrame([['Carl', 22],['Martha', 25],['Calvin', 12],['Stuart', 15]], columns=['Name', 'Age'])
B = pd.DataFrame([['Melvin', 25],['Martha', 34],['Lewis', 32],['Leo', 25]], columns=['Name', 'Age'])
```

```python
A
```

```python
B
```

```python
#1. Left Outer Join
# Left Outer Join: With a left outer join, all the records from 
# the first dataframe will be displayed, irrespective of whether
# the keys in the first dataframe can be found in the second dataframe. 
# Whereas, for the second dataframe, only the records with the keys 
# in the second dataframe that can be found in the first dataframe will be displayed.

df_left_outer = pd.merge(A, B, on='Name', how='left')
df_left_outer
```

```python
#2. Outer Join

# Full Outer Join: A full outer join returns all the rows from
# the left dataframe, all the rows from the right dataframe, 
# and matches up rows where possible, with NaNs elsewhere. 
# But if the dataframe is complete, then we get the same output.
# full outer join
df_full_outer = pd.merge(A, B, on='Name', how='outer')
df_full_outer
```

```python
#3. Inner Join

# Inner Join: Inner join is the most common type of join.
# It returns a dataframe with only those rows that have common characteristics. 
# This is similar to the intersection of two sets.
# inner join
df_inner = pd.merge(A, B, on='Name', how='inner')
df_inner
```

```python
#4. Right Outer Join

# Right Outer Join: For a right join, all the records from the 
# second dataframe will be displayed. However, only the records 
# with the keys in the first dataframe that can be found in the 
# second dataframe will be displayed.
df_right_outer = pd.merge(A, B, on='Name', how='right')
df_right_outer
```
