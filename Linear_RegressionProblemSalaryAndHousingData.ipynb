##15. For instance, if the â€˜priceâ€™ column consists of outliers, how can you make the data cleanand remove the redundancies?

```python
# normally we use IQR to sattel the Outliers 
# but z-score value and removing the less records with less than z-score will also work

Q1=df.price.quantile(0.25)
Q3=df.price.quantile(0.75)

IQR=Q3-Q1
df1=df
df= df[(df.price >= Q1-1.5*IQR) & (df.price <= Q3+1.5*IQR)]

sns.boxplot(df['price'])
plt.show()

#Answer here we have satteled outliers using IQR
#Correct option:   a. Calculate the IQR range and drop the values outside the range.
```
